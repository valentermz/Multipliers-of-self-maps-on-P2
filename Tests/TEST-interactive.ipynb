# `TEST` -  Decide whether a collection of rational numbers `u_i,v_i` is realizable by a quadratic map

## Introduction

In this script we test if there is any obstuction coming from our formulas for a collection of 14 rational numbers `u_i, v_i` to come from a quadratic self-map of the complex projective plane having a (forward) invariant line.

More precisely, let `[u_0,v_0,...,u_6,v_6]` be a collection of rational numbers and suppose there exists a quadratic map $f$ with non-degenerate fixed points $p_0,\ldots, p_6$, satisfying the following:

* `u_i, v_i` are the eigenvalues of $\mathrm{I}-Df(p_i)$,
* The points $p_0,\ldots,p_3$ lie away from the invariant line, and the points $p_4,p_5,p_6$ are on it,
* For i=4,5,6, the eigenvalue `u_i` corresponds to the direction tangent to the invariant line.

Then the coefficients `a,b,c_0,...,c_5` defining the map $f$ satisfy the equations given by `Eq, Rel, Form,` and `gc` given below. Therefore, the ideal `I` containing all these equations defines a non-empty variety (ie the ideal does not contain the constant 1).

After eliminating some auxiliary variables we obtain an ideal `J` of the subring generated by the variables `a,b,c_0,\ldots,c_5`. If the proposed spectra is realizable, we must have `J.dimension()` $\geq$ 0 (by convention the dimension of the empty variety is -1).

This test determines whether or not `J.dimension()` is non-negative.

### Usage

**Input:**  
Provide a list of 14 rational numbers as follows:
```python
Input = [u_0,v_0,u_1,v_1,u_2,v_2,u_3,v_3,u_4,v_4,u_5,v_5,u_6,v_6],
```
or, if the variable `'td_format'` is set to `'True'`, provide
```python
Input = [t_0,d_0,t_1,d_1,t_2,d_2,t_3,d_3,u_4,v_4,u_5,v_5,u_6,v_6],
```
where `t_i = u_i + v_i`, and `d_i = u_i*v_i`.

**Output:**  
The program will return either
```python
'Test PASSED'
```
if `J.dimension() != -1` (namely, if `J` defines a non-empty affine variety), or
```python
'Test FAILED'
```
otherwise.  
The final result also includes the dimension of (the variety defined by) `J` and information about the generators of `J`.

### Scope

A 'PASSED' result should be analyzed with care since the computations are carried over $\mathbb{Q}$ and not over $\mathbb{C}$.

The test is failed if and only if the constant 1 belongs to the ideal `J` (over $\mathbb{Q}$ and hence also over $\mathbb{C}$). In this case we can be certain that it is impossible for the values provided in `'Input'` to come from a quadratic map.

On the other hand, if the test is passed we only know that our equations do not provide an obstruction to realizability. Further analysis is needed to decide if the collection `'Input'` is truly realizable. However, a close look to the generators of the ideal `J` computed during the elimination process provides a system of equations for the variables `a,b,c_i` which are expected to be easily solved. Finding a solution to such system would confirm the realizability of the proposed spectra over $\mathbb{C}$ and would construct an explicit map that realizes such spectra.

**Expected output:**  
A generic realizable tuple defines a unique endomorphism $f$ that realizes it. Therefore, it defines an ideal `J` which is a maximal ideal in $\mathbb{C}[a,b,c_0,\ldots,c_5]$. Therefore, a generic realizable input will define an ideal `J` with the following properties:
* `J.dimension() = 0`,
* `J` is generated by 8 polynomials,
* Each of this generators is linear and involves only one variable from `(a,b,c_0,...,c_5)`.
